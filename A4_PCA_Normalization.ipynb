# PCA com e sem Normalização

O método PCA pressupõe que os atributos estão na mesma escala.   
Caso não estejam, é necessário fazer a normalização antes de aplicar o método.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
X = np.random.uniform(-1,1,size=(100,2))
X[:,0] = 1000*X[:,0]
X[:,1] = 0.001*X[:,0]+np.random.uniform(-0.3,0.3,100)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
opca = PCA(svd_solver='full')  # full = todos os atributos
X_pca = opca.fit_transform(X)

plt.rcParams['figure.figsize'] = [12, 6]
f, (ax1,ax2)  = plt.subplots(2,1)
ax1.set_aspect('equal')
ax1.set_ylim(-100,100)
ax1.scatter(X_pca[:,0],X_pca[:,1])

#ax2.set_aspect('equal')
ax2.set_ylim(-5,5)
ax2.scatter(X[:,0],X[:,1])
ax2.plot([0,100*opca.components_[0,0]],[0,100*opca.components_[0,1]],color='red')
ax2.plot([0,opca.components_[1,0]],[0,opca.components_[1,1]],color='red')
```

```python
scaler = StandardScaler()
X_std = scaler.fit_transform(X)

opca = PCA(svd_solver='full')
X_stdpca = opca.fit_transform(X_std)

plt.rcParams['figure.figsize'] = [12, 6]
f, (ax1,ax2)  = plt.subplots(2,1)
ax1.set_aspect('equal')
#ax1.set_ylim(-100,100)
ax1.scatter(X_stdpca[:,0],X_stdpca[:,1])

ax2.set_aspect('equal')
ax2.scatter(X_std[:,0],X_std[:,1])
ax2.plot([0,opca.components_[0,0]],[0,opca.components_[0,1]],color='red')
ax2.plot([0,opca.components_[1,0]],[0,opca.components_[1,1]],color='red')
```
